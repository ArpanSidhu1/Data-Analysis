```python
# Question - 1 : what percentage of all licensed sets ever released were star wars themed store the answer in the form of float in the _force variable. ? 
# Question -  2 : In which Year does Star wars was not the most popular licensed theme (in terms of number of sets released that year) and store it in the new_Era variable ?
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
theme = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
print(theme.head(5))
df.head(5) # so there were 2 datasets with one common attribute but in that attribute both the columns name were different
```

```python
theme=theme.rename(columns={'name':"parent_theme"}) # rename() - attribute columns={'name' : 'parent_theme'}
print(theme.head(5))  # changed Second data set column 
```

```python
merged_data=df.merge(theme,on=["parent_theme"]) # it merged them into a single data set. dataset1.merge(dataset2,on=["parent_theme"])
#merged_data.to_csv("D:\\merge.csv")
```

```python
merged_data[merged_data["set_num"].isnull()].shape
```

```python
licensed = merged_data[merged_data.is_licensed==1]
licensed = licensed.dropna(subset=["set_num"])  # the drop na will only work on set num column if we had simply use drop na then  it will remove every row checking all the dataset which is containing atleast one missing value.

starwars=licensed[licensed["parent_theme"]=="Star Wars"] # This DataSet Contains all the star wars theme which was licensed.
#starwars.to_csv("D:\\starlicense.csv")
print(starwars.shape)
print(licensed.shape)
the_force=round(starwars.shape[0]*100/licensed.shape[0],2)
```

```python
# Answer of First Question
the_force
```

```python
licensed_sort=licensed.sort_values("year")
licensed_sort["count"]=1

sum_df = licensed_sort.groupby(by=["year","parent_theme"]).sum().reset_index()
#sum_df.tail(25)
max_df= sum_df.sort_values('count',ascending=False).drop_duplicates(['year'])
print(max_df.sort_values('year'))
```
